## Overview / Abstract

This project investigates whether court surface (clay, grass, hard) is associated with differences in first-serve and second-serve win percentages in 2024 men's singles Grand Slam matches. Using match-level data from the ATP tour (Jeff Sackmann's `tennis_atp` repository), we cleaned the 2024 season data down to 485 Grand Slam matches and reshaped them into 970 player-level serve observations for Kruskal-Wallis tests with post-hoc pairwise comparisons. We found that first-serve win percentage differs significantly across surfaces (H = 61.21, p < 0.001), with grass courts showing the highest mean (74.5%) and clay the lowest (67.9%). Second-serve win percentage also varies by surface (H = 14.25, p < 0.001), but with a smaller effect size, consistent with our hypothesis that surface is more strongly associated with first-serve outcomes than second-serve outcomes. These results matter because they show that even within elite competition, surface conditions are associated with measurable differences in how much advantage players gain from serve quality. Because surface and tournament identity are partially confounded in this single-season design, the findings should be interpreted as descriptive associations rather than causal effects.

## Research Question

To what degree is court surface (clay, grass, hard) associated with differences in first-serve and second-serve win percentage in 2024 men's singles Grand Slam matches?

This is an associational study using observational data; we examine whether serve outcomes differ across surfaces, but we cannot establish a causal relationship because surface is not randomly assigned and is partially confounded with tournament identity, player draw, and match conditions.

## Background and Prior Work

Court surface plausibly shapes serve outcomes because bounce height, skid, and pace influence how quickly the returner can react after the serve. Grand Slams provide a useful comparison set for this question because the 2024 men's singles calendar includes hard courts (Australian Open and US Open), clay (Roland Garros), and grass (Wimbledon) at the same elite competitive level.

Existing work motivates our choice of serve metrics. Kokta's modeling project argues that first-serve and second-serve points won are among the most informative serving statistics for match prediction, which supports using those point-winning measures as our core outcomes.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Whiteside et al. show that professional serve outcomes depend on multiple spatial and tactical attributes rather than raw speed alone, which reinforces the value of studying overall point success instead of aces in isolation.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) The tennis commentary source makes a similar practical point: aces are visible, but broader serve-point metrics better capture serve dominance across matches.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

What these sources do not provide is a focused, single-season comparison of first-serve and second-serve win percentage across all three Grand Slam surface types in one analysis. Our contribution is therefore narrower and descriptive: we compare 2024 men's Grand Slam serve outcomes across clay, grass, and hard courts using a common match-statistics source and a consistent player-level reshaping procedure.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

## Hypothesis


We predict that court surface will be associated with serve win percentages, with the highest first-serve win percentages occurring on grass and the lowest on clay, while second-serve win percentages show smaller differences across surfaces than first-serve win percentages. Because first serves are typically struck more aggressively and second serves prioritize stability, we expect surface to matter more for first-serve point outcomes than for second-serve point outcomes. This directional expectation is also consistent with general court-pace differences, where grass tends to reward low, skidding serves more than clay.

Key confounds to acknowledge: surface is partially confounded with tournament identity (grass = Wimbledon only, clay = Roland Garros only, hard = Australian Open + US Open), and the analysis is limited to a single season (2024) with a specific player mix in each draw. Any observed surface associations may partly reflect tournament-specific conditions rather than surface alone.

### ATP Men’s Tennis Matches – 2024 Season 

The ATP Matches 2024 dataset contains detailed match-level information for all ATP men’s professional tennis matches played between January 1 and December 31, 2024. The data comes from Jeff Sackmann’s tennis_atp repository and includes 3,076 matches and 49 variables. Each row represents a single tennis match and each column represents one variable, meaning the dataset is already in tidy format.

The dataset includes several important categories of variables. Tournament information includes surface (Hard, Clay, Grass), tournament level (Grand Slam, Masters 1000, ATP 250/500), and date. Player information includes winner and loser names, handedness, height (cm), and age (years). Match outcome variables include score, round, best_of (3 or 5 sets), and match duration in minutes.

Several performance metrics are particularly important:

Aces (w_ace, l_ace): Number of serves that result in an immediate point. Measured as a count per match. Professional players typically average 5–15 aces per match. Higher ace counts generally indicate stronger serving performance.

Double Faults (w_df, l_df): Number of missed second serves. Measured as a count per match. Professionals usually aim to stay under 3–5 per match since double faults directly give points to the opponent.

Match Duration (minutes): Measured in minutes. Best-of-three matches typically last 90–150 minutes; best-of-five matches may exceed 3–5 hours.

ATP Ranking (winner_rank, loser_rank): Numerical ranking (1 = best). Rankings are based on a rolling 52-week point system. Lower numbers indicate stronger players.

Dataset Concerns:

Approximately 7% of all values are missing. Missingness is not random. It is strongly associated with incomplete matches (retirements and walkovers). There are 101 matches marked as walkover (W/O) or retirement (RET), and 238 matches missing duration data. Around 60 matches are missing serve statistics such as aces and double faults. These missing values are systematic and expected because incomplete matches do not generate full statistics.

There are also extreme match duration values (e.g., 0 minutes for walkovers and matches exceeding 220 minutes), but these represent legitimate real-world scenarios and should not automatically be removed.

## Data Cleaning

Several targeted cleaning steps were applied before analysis:

1. **Removed incomplete matches:** Retirements (RET) and walkovers (W/O) were excluded because they produce incomplete or misleading serve totals.
2. **Computed serve win percentages:** First-serve win % = first-serve points won / first serves in. Second-serve win % = second-serve points won / second-serve attempts (total serve points minus first serves in).
3. **Filtered invalid rows:** Matches with missing surfaces, non-positive serve denominators, or impossible derived percentages outside [0, 1] were removed.
4. **Reshaped to player level:** Each cleaned Grand Slam match was split into one winner row and one loser row so the analysis compares individual serve performances rather than only match winners.

Of the 508 Grand Slam matches in the raw file, 22 were retirements or walkovers and one additional Roland Garros match was excluded because the recorded second-serve totals implied a win percentage above 100%, which is not physically possible. After cleaning, hard-court observations total 484 player-level rows, compared with 244 on clay and 242 on grass.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("data/00-raw/atp_matches_2024.csv")
print("Dataset shape:", df.shape)
display(df.head(3))

required_cols = [
    "surface",
    "w_svpt", "w_1stIn", "w_1stWon", "w_2ndWon",
    "l_svpt", "l_1stIn", "l_1stWon", "l_2ndWon",
    "score", "tourney_level", "tourney_name"
]
missing_required = [c for c in required_cols if c not in df.columns]
if missing_required:
    raise KeyError(f"Missing expected columns: {missing_required}")

cols_of_interest = [
    "tourney_name", "tourney_date", "surface",
    "winner_name", "loser_name",
    "w_svpt", "w_1stIn", "w_1stWon", "w_2ndWon",
    "l_svpt", "l_1stIn", "l_1stWon", "l_2ndWon",
    "score", "minutes"
]
print()
print("Sample rows for hypothesis-related columns:")
display(df[cols_of_interest].head(8))

surface_counts = df["surface"].value_counts(dropna=False)
surface_perc = df["surface"].value_counts(dropna=False, normalize=True) * 100
surface_dist = pd.DataFrame({"count": surface_counts, "percent": surface_perc.round(2)})
print()
print("Court surface distribution (counts + percents):")
display(surface_dist)

ax = surface_counts.dropna().plot(kind="bar", title="Match Counts by Court Surface")
ax.set_xlabel("Surface")
ax.set_ylabel("Number of matches")
plt.tight_layout()
plt.show()

interest_cols = ["surface", "w_svpt", "w_1stIn", "w_1stWon", "w_2ndWon", "l_svpt", "l_1stIn", "l_1stWon", "l_2ndWon"]
missing_interest = df[interest_cols].isna().sum().to_frame("missing_count")
missing_interest["missing_percent"] = (missing_interest["missing_count"] / len(df) * 100).round(2)
print()
print("Missingness for variables of interest:")
display(missing_interest)

print()
print("Basic summary stats (raw counts) for serve-related columns:")
display(df[interest_cols[1:]].describe().T)

analysis_df = df.copy()
analysis_df["is_ret_wo"] = analysis_df["score"].astype(str).str.contains(r"RET|W/O", regex=True, na=False)
analysis_df["w_2ndIn"] = analysis_df["w_svpt"] - analysis_df["w_1stIn"]
analysis_df["l_2ndIn"] = analysis_df["l_svpt"] - analysis_df["l_1stIn"]
analysis_df["w_1st_win_pct"] = analysis_df["w_1stWon"] / analysis_df["w_1stIn"]
analysis_df["w_2nd_win_pct"] = analysis_df["w_2ndWon"] / analysis_df["w_2ndIn"]
analysis_df["l_1st_win_pct"] = analysis_df["l_1stWon"] / analysis_df["l_1stIn"]
analysis_df["l_2nd_win_pct"] = analysis_df["l_2ndWon"] / analysis_df["l_2ndIn"]

pct_cols = ["w_1st_win_pct", "w_2nd_win_pct", "l_1st_win_pct", "l_2nd_win_pct"]
ret_wo_count = int(analysis_df["is_ret_wo"].sum())
clean_mask = (
    analysis_df["surface"].notna() &
    (~analysis_df["is_ret_wo"]) &
    (analysis_df["w_1stIn"] > 0) &
    (analysis_df["w_2ndIn"] > 0) &
    (analysis_df["l_1stIn"] > 0) &
    (analysis_df["l_2ndIn"] > 0)
)
pre_bounds_df = analysis_df.loc[clean_mask].copy()
analysis_df = pre_bounds_df.dropna(subset=pct_cols).copy()
for c in pct_cols:
    analysis_df = analysis_df[(analysis_df[c] >= 0) & (analysis_df[c] <= 1)]

invalid_pct_rows = pre_bounds_df.loc[~pre_bounds_df.index.isin(analysis_df.index), [
    "tourney_name", "surface", "winner_name", "loser_name", "score"
] + pct_cols]

cleaning_summary = pd.DataFrame([
    ["Raw ATP 2024 matches", len(df)],
    ["RET/W/O matches excluded", ret_wo_count],
    ["Cleaned ATP matches retained", len(analysis_df)],
    ["Raw Grand Slam matches", int((df["tourney_level"] == "G").sum())],
    ["Cleaned Grand Slam matches retained", int((analysis_df["tourney_level"] == "G").sum())],
])
cleaning_summary.columns = ["Stage", "Observations"]
print()
print("Cleaning summary:")
display(cleaning_summary)

if not invalid_pct_rows.empty:
    print()
    print("Rows removed because derived serve percentages were impossible:")
    display(invalid_pct_rows)

print()
print("After cleaning:")
print("Clean analysis_df shape:", analysis_df.shape)
print("Surfaces in cleaned data:")
display(analysis_df["surface"].value_counts().to_frame("count"))

print()
print("Overall summary stats for serve win % variables:")
display(analysis_df[pct_cols].describe().T)

analysis_path = "data/02-processed/atp_matches_2024_analysis.csv"
try:
    from pathlib import Path
    Path("data/02-processed").mkdir(parents=True, exist_ok=True)
    analysis_df.to_csv(analysis_path, index=False)
    print()
    print("Saved cleaned analysis dataset to:", analysis_path)
except Exception as e:
    print()
    print("(Info) Could not save to disk in this environment:", e)
```

## Results

### Exploratory Data Analysis

In the data wrangling section above, we loaded the full ATP 2024 match dataset (3,076 matches), removed incomplete matches (retirements and walkovers), computed first-serve and second-serve win percentages, and filtered out rows with missing serve statistics. This produced a cleaned dataset of complete matches with valid serve metrics.

For the EDA, we focus on **2024 Grand Slam matches only**, consistent with our research question. We reshape each match into two **player-level observations** (one for the winner and one for the loser) so that every player's serve performance in each match is analyzed independently. This means a match between Player A and Player B produces two rows — one for each player's serve statistics — giving us approximately twice as many observations as matches. This reshaping is done transparently in the code cell below. Hard-court observations are about double the size of clay or grass because two Grand Slam tournaments are played on hard courts.

```python
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")
plt.rcParams['figure.dpi'] = 120
plt.rcParams['figure.figsize'] = (10, 6)

grand_slam_df = analysis_df[analysis_df['tourney_level'] == 'G'].copy()
print(f"Grand Slam matches after cleaning: {len(grand_slam_df)}")
print()
print("Tournament breakdown:")
display(grand_slam_df['tourney_name'].value_counts().to_frame('matches'))

winner_stats = grand_slam_df[[
    'surface', 'tourney_name', 'w_1st_win_pct', 'w_2nd_win_pct',
    'w_ace', 'w_df', 'w_svpt', 'w_1stIn'
]].copy()
winner_stats.columns = [
    'surface', 'tournament', 'first_serve_win_pct', 'second_serve_win_pct',
    'aces', 'double_faults', 'serve_points', 'first_serves_in'
]
winner_stats['outcome'] = 'Winner'

loser_stats = grand_slam_df[[
    'surface', 'tourney_name', 'l_1st_win_pct', 'l_2nd_win_pct',
    'l_ace', 'l_df', 'l_svpt', 'l_1stIn'
]].copy()
loser_stats.columns = [
    'surface', 'tournament', 'first_serve_win_pct', 'second_serve_win_pct',
    'aces', 'double_faults', 'serve_points', 'first_serves_in'
]
loser_stats['outcome'] = 'Loser'

eda_df = pd.concat([winner_stats, loser_stats], ignore_index=True)
eda_df['ace_rate'] = eda_df['aces'] / eda_df['serve_points']
eda_df['df_rate'] = eda_df['double_faults'] / eda_df['serve_points']

print()
print(f"Expected player-level observations (2 per cleaned Grand Slam match): {len(grand_slam_df) * 2}")
print(f"Actual player-level observations analyzed: {len(eda_df)}")
print()
print("Breakdown by surface:")
display(eda_df['surface'].value_counts().to_frame('player_performances'))
```

First-serve win percentages have a tighter distribution, reflecting the inherent advantage of a well-placed first serve. A few outliers are visible at the tails — particularly some first-serve win percentages below 45% and second-serve win percentages below 25%, likely representing matches where a player struggled heavily on serve. ### Serve Win Percentage by Court Surface

Our central question asks whether court surface is associated with first-serve and second-serve win percentages differently. We compare both metrics across clay, grass, and hard courts using boxplots and summary statistics.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
surface_order = ['Clay', 'Grass', 'Hard']
palette = {'Clay': '#D2691E', 'Grass': '#228B22', 'Hard': '#4169E1'}

# First-serve win% by surface
sns.boxplot(ax=axes[0], x='surface', y='first_serve_win_pct', data=eda_df,
            order=surface_order, hue='surface', palette=palette, legend=False)
axes[0].set_title('First-Serve Win % by Surface', fontweight='bold', fontsize=13)
axes[0].set_xlabel('Court Surface')
axes[0].set_ylabel('First-Serve Win %')
axes[0].set_ylim(0.3, 1.0)

# Second-serve win% by surface
sns.boxplot(ax=axes[1], x='surface', y='second_serve_win_pct', data=eda_df,
            order=surface_order, hue='surface', palette=palette, legend=False)
axes[1].set_title('Second-Serve Win % by Surface', fontweight='bold', fontsize=13)
axes[1].set_xlabel('Court Surface')
axes[1].set_ylabel('Second-Serve Win %')
axes[1].set_ylim(0.1, 0.85)

plt.tight_layout()
plt.show()

# Summary statistics table
print("Summary Statistics: Serve Win % by Surface\n")
summary = eda_df.groupby('surface')[['first_serve_win_pct', 'second_serve_win_pct']].agg(
    ['count', 'mean', 'median', 'std']
).round(4)
display(summary)
```

The boxplots suggest that first-serve win percentage varies noticeably across surfaces. Grass courts appear to have the highest median first-serve win percentage, followed by hard courts, with clay showing the lowest. This is consistent with the hypothesis that faster surfaces reward aggressive first serves more than slower surfaces such as clay, where returners generally have more time after the bounce.

For second-serve win percentage, the differences across surfaces are smaller than for first-serve win percentage, but they are not flat. The medians are relatively close, with grass slightly higher and hard/clay lower. This pattern is consistent with our hypothesis that second-serve performance varies less by surface than first-serve performance, since second serves rely more on placement and spin than on raw pace.

**Sample sizes:** Clay has 244 player-level observations, Grass has 242, and Hard has 484. The boxplots also reveal outliers on all three surfaces, representing a handful of player-performances with unusually low serve-point win rates.

**Note:** These are descriptive comparisons, not formal hypothesis tests. The Statistical Analysis section below tests whether the observed gaps are large enough relative to within-surface variation to count as statistically significant in this sample.

### Serve Accuracy by Surface

To complement serve win percentages (effectiveness), we examine serve accuracy/consistency by surface using two rates: first-serve in rate and second-serve in rate. This helps separate whether surface differences come from getting serves in more often, or from winning more points once serves land.

```python
# Compute serve-accuracy metrics with safe denominators
eda_df['second_serve_attempts'] = eda_df['serve_points'] - eda_df['first_serves_in']
eda_df.loc[eda_df['second_serve_attempts'] < 0, 'second_serve_attempts'] = np.nan

eda_df['first_serve_in_rate'] = np.where(
    eda_df['serve_points'] > 0,
    eda_df['first_serves_in'] / eda_df['serve_points'],
    np.nan
)

eda_df['second_serve_in_rate'] = np.where(
    eda_df['second_serve_attempts'] > 0,
    (eda_df['second_serve_attempts'] - eda_df['double_faults']) / eda_df['second_serve_attempts'],
    np.nan
)

for col in ['first_serve_in_rate', 'second_serve_in_rate']:
    eda_df[col] = eda_df[col].where(eda_df[col].between(0, 1), np.nan)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# First-serve in rate by surface
sns.boxplot(ax=axes[0], x='surface', y='first_serve_in_rate', data=eda_df,
            order=surface_order, hue='surface', palette=palette, legend=False)
axes[0].set_title('First-Serve In Rate by Surface', fontweight='bold', fontsize=13)
axes[0].set_xlabel('Court Surface')
axes[0].set_ylabel('First-Serve In Rate')
axes[0].set_ylim(0.3, 1.0)
# Second-serve in rate by surface
sns.boxplot(ax=axes[1], x='surface', y='second_serve_in_rate', data=eda_df,
            order=surface_order, hue='surface', palette=palette, legend=False)
axes[1].set_title('Second-Serve In Rate by Surface', fontweight='bold', fontsize=13)
axes[1].set_xlabel('Court Surface')
axes[1].set_ylabel('Second-Serve In Rate')
axes[1].set_ylim(0.4, 1.02)

plt.tight_layout()
plt.show()

print('Summary Statistics: Serve Accuracy by Surface\n')
accuracy_summary = eda_df.groupby('surface')[[
    'first_serve_in_rate', 'second_serve_in_rate'
]].agg(['count', 'mean', 'median', 'std']).round(4)
display(accuracy_summary)
```

Serve accuracy varies less across surfaces than first-serve win percentage. First-serve in rate shows only modest surface differences, while second-serve in rate remains high and tightly clustered across all surfaces.

Combined with the earlier win-percentage boxplots, this suggests that surface effects on first serve are driven more by the reward after the serve lands than by large differences in first-serve landing frequency. For second serves, both in-rate and win-rate are comparatively stable, consistent with smaller surface sensitivity.

**Note:** Rows with zero second-serve attempts are assigned `NaN` for second-serve in rate to avoid division-by-zero artifacts.

### Statistical Analysis

The EDA above suggests that first-serve win percentage differs across surfaces more than second-serve win percentage. To determine whether these differences are statistically meaningful, we conduct formal hypothesis tests.

**Approach:** We use the Kruskal-Wallis H-test, a non-parametric alternative to one-way ANOVA, to test whether the distribution of serve win percentages differs across the three surface groups (Clay, Grass, Hard). We use Kruskal-Wallis because the groups have unequal sample sizes and the serve metrics do not need to satisfy strict normality assumptions for this comparison.

For significant omnibus tests, we follow up with pairwise Mann-Whitney U tests with Bonferroni correction for the three surface comparisons (alpha = 0.05 / 3 = 0.0167). We report the rank-biserial correlation *r* as an effect size measure, where |r| < 0.1 is negligible, 0.1-0.3 is small, 0.3-0.5 is medium, and > 0.5 is large.

One caution is important for interpretation: each match contributes two player-level rows, so the observations are not fully independent. These tests therefore provide evidence of association in this sample, not a fully isolated estimate of a causal surface effect.

As a robustness check, we also compare serve outcomes by individual tournament. That check helps show whether the pattern is concentrated in a single event, but it does not resolve the broader confounding between surface and tournament identity.

```python
from scipy import stats
from itertools import combinations

surfaces = ['Clay', 'Grass', 'Hard']
alpha = 0.05
bonferroni_alpha = alpha / 3

print("Note: each match contributes two player-level observations, so these tests should be interpreted as evidence of association rather than a fully independent causal estimate.")
print("=" * 70)
print("OMNIBUS TESTS: Kruskal-Wallis H-test")
print("=" * 70)

kw_results = {}
for metric, label in [('first_serve_win_pct', 'First-Serve Win %'),
                      ('second_serve_win_pct', 'Second-Serve Win %')]:
    groups = [eda_df.loc[eda_df['surface'] == s, metric].values for s in surfaces]
    H_stat, p_val = stats.kruskal(*groups)
    n = len(eda_df)
    k = len(surfaces)
    eta_sq = (H_stat - k + 1) / (n - k)
    kw_results[metric] = {'H': H_stat, 'p': p_val, 'eta_sq': eta_sq}

    sig_label = "***" if p_val < 0.001 else ("**" if p_val < 0.01 else ("*" if p_val < 0.05 else "ns"))
    print()
    print(f"{label}:")
    print(f"  H = {H_stat:.3f}, p = {p_val:.6f} {sig_label}")
    print(f"  Effect size (eta^2) = {eta_sq:.4f}")

print()
print("=" * 70)
print(f"POST-HOC: Pairwise Mann-Whitney U tests (Bonferroni alpha = {bonferroni_alpha:.4f})")
print("=" * 70)

posthoc_rows = []
for metric, label in [('first_serve_win_pct', 'First-Serve Win %'),
                      ('second_serve_win_pct', 'Second-Serve Win %')]:
    print()
    print(f"{label}:")
    for s1, s2 in combinations(surfaces, 2):
        g1 = eda_df.loc[eda_df['surface'] == s1, metric]
        g2 = eda_df.loc[eda_df['surface'] == s2, metric]
        u_stat, p_val = stats.mannwhitneyu(g1, g2, alternative='two-sided')
        r_effect = 1 - (2 * u_stat) / (len(g1) * len(g2))
        sig = "Yes" if p_val < bonferroni_alpha else "No"

        posthoc_rows.append({
            'Metric': label,
            'Comparison': f'{s1} vs {s2}',
            'U statistic': f'{u_stat:.0f}',
            'p-value': f'{p_val:.6f}',
            'Effect size (r)': f'{r_effect:.3f}',
            'Significant': sig
        })
        print(f"  {s1} vs {s2}: U = {u_stat:.0f}, p = {p_val:.6f}, r = {r_effect:.3f} -> {sig}")

print()
print("=" * 70)
print("SUMMARY TABLE")
print("=" * 70)
posthoc_df = pd.DataFrame(posthoc_rows)
display(posthoc_df)

print()
print("=" * 70)
print("ROBUSTNESS CHECK: Per-Tournament Kruskal-Wallis")
print("=" * 70)

tourney_names = sorted(eda_df['tournament'].unique())
for metric, label in [('first_serve_win_pct', 'First-Serve Win %'),
                      ('second_serve_win_pct', 'Second-Serve Win %')]:
    groups_t = [eda_df.loc[eda_df['tournament'] == t, metric].values for t in tourney_names]
    H_t, p_t = stats.kruskal(*groups_t)
    print()
    print(f"{label} (by tournament):")
    print(f"  Tournaments: {tourney_names}")
    print(f"  H = {H_t:.3f}, p = {p_t:.6f}")

print()
print("=" * 70)
print("GROUP MEANS BY SURFACE")
print("=" * 70)
means_table = eda_df.groupby('surface')[['first_serve_win_pct', 'second_serve_win_pct']].agg(['mean', 'std', 'count']).round(4)
display(means_table)
```

### Statistical Analysis Results

#### First-Serve Win Percentage

The Kruskal-Wallis test confirms that first-serve win percentage differs significantly across surfaces (H = 61.21, p < 0.001), with a small-to-medium effect size (eta^2 = 0.061). Post-hoc pairwise comparisons (Bonferroni-corrected alpha = 0.0167) show:

| Comparison | Significant? | Effect Size (r) | Interpretation |
|---|---|---|---|
| Clay vs Grass | Yes (p < 0.001) | 0.381 (medium) | Grass higher than Clay |
| Clay vs Hard | Yes (p < 0.001) | 0.284 (small) | Hard higher than Clay |
| Grass vs Hard | Yes (p = 0.008) | -0.121 (small) | Grass modestly higher than Hard |

All three pairwise comparisons are significant for first-serve win percentage. The largest gap is between Clay (mean 67.9%) and Grass (mean 74.5%), which is consistent with the idea that faster surfaces are associated with stronger first-serve outcomes.

#### Second-Serve Win Percentage

The Kruskal-Wallis test also shows a significant difference for second-serve win percentage (H = 14.25, p < 0.001), but with a much smaller effect size (eta^2 = 0.013). Post-hoc results are more mixed:

| Comparison | Significant? | Effect Size (r) | Interpretation |
|---|---|---|---|
| Clay vs Grass | Yes (p = 0.004) | 0.150 (small) | Grass higher than Clay |
| Clay vs Hard | No (p = 0.723) | -0.016 (negligible) | No meaningful difference |
| Grass vs Hard | Yes (p < 0.001) | -0.167 (small) | Grass modestly higher than Hard |

Second-serve win percentage therefore shows weaker and less consistent surface differences than first-serve win percentage. Two of the three pairwise comparisons survive Bonferroni correction, and all of the estimated effects are small.

#### Robustness Check

When the comparison is run at the tournament level (Australian Open, Roland Garros, Wimbledon, US Open), first-serve and second-serve win percentages still differ significantly across events. That result is useful as a stability check, but it should not be read as separating tournament effects from surface effects. Tournament identity remains the main confound because each non-hard surface is represented by only one Grand Slam event.

#### Key Takeaway

The formal tests support the descriptive pattern from the EDA: court surface is strongly associated with first-serve win percentage and more modestly associated with second-serve win percentage in this 2024 Grand Slam sample. The evidence is strongest for first serves, where grass consistently outperforms clay and hard courts in both descriptive and inferential comparisons.

## Ethics & Privacy

This project uses publicly available professional match statistics, so privacy risk is minimal and no sensitive personal information is collected or inferred. The more important ethical issue is interpretation: these results describe a narrow population of elite men's singles Grand Slam matches from one season, not all tennis players or all competitive contexts. It would be inappropriate to generalize directly to women's tennis, lower-level tournaments, amateur play, or future seasons without additional evidence.

A second ethical consideration is avoiding overclaiming. Because the design is observational and surface is partially confounded with tournament identity, player draw, and playing conditions, the notebook should treat the findings as associations rather than proof that surface alone causes the observed differences. The results are best used as a careful descriptive summary of 2024 Grand Slam serve outcomes, not as a basis for judging individual players too strongly or making high-stakes decisions such as betting claims.

## Discussion and Conclusion

This project asked whether court surface (clay, grass, hard) is associated with differences in first-serve and second-serve win percentages in 2024 men's singles Grand Slam matches. Using 970 player-level observations from 485 cleaned Grand Slam matches, we found strong evidence that surface is associated with first-serve win percentage and weaker but still statistically detectable differences for second-serve win percentage. These results are consistent with the hypothesis that faster surfaces reward aggressive serving more strongly than slower ones.

Our results connect back to prior work in two ways. First, they reinforce the idea that first-serve and second-serve points won are informative performance measures, rather than relying on aces alone. Second, they add a surface-specific comparison within one season of elite play: first-serve win percentage on grass (mean 74.5%) was roughly 6.6 percentage points higher than on clay (67.9%), while second-serve win percentage showed only small differences across surfaces. The supporting analysis of ace rates and serve accuracy points in the same direction: ace rates rise on faster courts, while first-serve in-rate changes much less, suggesting the main difference occurs after the serve lands.

There are several important limitations. Surface and tournament are partially confounded: grass means Wimbledon, clay means Roland Garros, and hard means the Australian Open plus the US Open. Any observed surface association may therefore partly reflect tournament-specific conditions such as weather, ball type, altitude, or draw composition. In addition, the player-level reshaping creates two observations per match, so the independence assumption behind the statistical tests is only approximate. Finally, the analysis covers one season of men's singles Grand Slam main-draw matches, so the findings should not be generalized beyond that scope.

Within those limits, the notebook provides a clear descriptive result: court surface is meaningfully associated with first-serve effectiveness in 2024 men's Grand Slam tennis, while the association for second-serve effectiveness is smaller and less consistent. Natural next steps would be to extend the analysis across multiple seasons, add women's Grand Slam data for comparison, and fit mixed-effects models that account for repeated players and match-level pairing.
